Now repeat the previous exercise on computer. 

Choose at least one stopping criterion and its parameter, and the bail-out number of iterations, R. Implement the above as an algorithm that at each step prints out the x value, the function's value f(x) and gradient rf(x). At the end it should print the solution, and also how many iterations it took for the loop to converge (to finish).

```python
# function f = -x^2 and its gradient -2x for 1-Dimensional Gradient Ascent
def f(x):
    return (-1)*x*x
def grad(x):
    return (-2)*x
```

```python
def gradient_ascent_1d(x0, R, max_iterations, epsilon=0.0001):
    """
    Function
    --------
    gradient_ascent_1d: implement 1-dimensional gradient ascent and print x, fx, gradient of x, number of iterations and
    maximum of fx

    Parameters
    ----------
    x0 : int, Value of x0
    R: float, Learning rate
    max_iterations: Int, maximum number of iterations
    """
    for i in range(max_iterations):
        #print ("x value:",x0)
        #print ("Function value:", f(x0))
        #print ("Gradient of the function", grad(x0),'\n')
        x1 = x0 + (R*grad(x0))
        
        #stopping condition
        if(abs(x1-x0) <=epsilon):
            print("Converged\n")
            break
        x0 = x1
    print("Number of Iterations:", i)
    print("Maximum of the function:", x1)
```

#### Answer:
As we increase the learning rate, the number of iterations to converge also increases.
As we make the value of epsilon smaller(stopping criterion, difference between current value of x and previous value of x in this case), number of iterations to converge also increases.

#### Question 7

Now code the algorithm above for 2D case. Do it in matrix form! As before, experiment with a few different learning rates and see how does this influence the speed of convergence. Comment and explain your findings.

```python
# function fx and its gradient for N-dimensional Gradient Ascent

def f_nd(x, A):
    return (-1)*x.T @ A @ x

def grad_nd(x, A):
    return (-1)*(A + A.T) @ x 
```

```python
def gradient_ascent_nd(x0, A, R, max_iterations, xvalues, epsilon=0.000001):
    """
    Function
    --------
    gradient_ascent_nd: implement N-dimensional gradient ascent and print x, fx, gradient of x, number of iterations and
    maximum of fx

    Parameters
    ----------
    x0 : int, Value of x0
    A : matrix, 2 x 2 symmetric matrix
    R: float, Learning rate
    max_iterations: Int, maximum number of iterations
    xvalues: matrix, initial value of x0
    epsilon: stopping criteria for convergence
    """
    start_time = time.time()
    for i in range(max_iterations):
        #print ("x value:",x0)
        #print ("Gradient of the function", grad(x0, A),'\n')
        #Displaying first 10 value of f(x)
        if (i<=10):
            print("Function of x:", f_nd(x0, A))
        x1 = x0 + (R*grad_nd(x0, A))
        #print("Function of x1:", x1)
        xvalues = np.vstack((xvalues, x1.T))
        #stopping condition
        if(abs(x1-x0).any() <=epsilon):
            print ("Converged")
            break
        x0 = x1
    print("\nIterations:", i)
    print("\nMaximum of the function:", x1)
    print("\nTime taken:", time.time() - start_time)
    print("\nAll values of x:", xvalues)
    return xvalues
```

```python
x = np.matrix("2;-3")
print(x)
A = np.matrix("1,2; 2,8")
print(A)
```

```python
saved_x1 = gradient_ascent_nd(x, A, 0.1, 10000, xvalues=x.T)
```

```python
saved_x2 = gradient_ascent_nd(x, A, 0.01, 10000, xvalues=x.T)
```

#### Answer:
For the same value of x and different values of learning rate, we see that the number of iterations and subsequently the time taken to converge is more for smaller learning rate. 

#### Question 8

Visualize your algorithm's work:Hint: look at the example notebook code.

Mark all the intermittend points (the x1-s) on the plot and connect these with lines.
Note: you may want to save, instead of print, these values in your code now. In terms of

Add the function contours on the plot. In R you can use graphics::contour or ggplot2::geom_contour.
In matplotlib plt.contour does a similar job.
Note: try to use fixed 1:1 aspect ratio. Otherwise the direction of gradient does not coincide
with that of the steepest ascent on the figure.

Explain what you see.

```python
def gradient_ascent_contour(xvalues, A, max_iterations, lev, param1=5, param2=5):
    """Plot contour for N-dimensional Gradient Ascent"""
    ## create 30x30 grid matrix
    n = 30
    ex1 = np.linspace(-param1, param1, num=n)
    ex2 = np.linspace(-param2, param2, num=n)
    grid1, grid2 = np.meshgrid(ex1, ex2)
    ## fill the grid via looping.  You may prefer to use ufuncs instead.
    z = np.empty_like(grid1)
    for i in range(z.shape[0]):
        for j in range(z.shape[1]):
            x = np.matrix([grid1[i,j], grid2[i,j]]).T
            z[i,j] = f_nd(x, A)

    ## Make the plot
    plt.figure(figsize=(9,9))
    p = plt.contour(grid1, grid2, z,
                    levels = lev,
                    colors = 'black')
    plt.clabel(p, inline=1, fontsize=10)     
    xvalues = np.array(xvalues)
    maxval = min(100, len(xvalues))
    for i in range(maxval):
        plt.scatter(xvalues[i][0],xvalues[i][1], c='red', s=50)
        if(i>0):
            plt.plot([xvalues[i][0],xvalues[i-1][0]],[xvalues[i][1],xvalues[i-1][1]], c='red')
    plt.scatter(xvalues[-1][0],xvalues[-1][1], c='blue', s=50)
    plt.show()
```

```python
levels = [-50, -40, -30, -20, -10, -5, 0, 5, 10, 20, 30, 40, 50]
gradient_ascent_contour(saved_x1, A, 10000, levels, param1=5, param2=5)
```

#### Answer:
For each subsequent value of x, the factor by which the value of x decreases also becomes smaller. The difference between the points that are near the origin cannot be clearly seen as these points are very close. They achieve convergence at the origin. The blue point at the origin is the last point.

### 2. Outliers in different distributions

#### 2.1 Normal Distribution

Pick your number of repetitions R (1000 is a good choice), and sample sizes N (10, 1000, 100,000 are good choices, but you may want to adjust if the latter is too slow).

For R times create sample of N of standard normals, and take the mean of it. You'll have R means of N standard normals.
Compute the mean-of-means, it's standard deviation, and 95% confidence region.

The 95% confidence region is the opposite of computing the percentage of observations that fit into the interval. The easiest way to find it is using sample quantiles. Check out the quantile function in R, or np.percentile function in python. For instance, np.percentile(x, 10) will give the 10th percentile q10 of an array x. 10th percentile is such a value that 10% of the sample values are smaller than that value.

Repeat the above with three different (and very different) N-s, such as 10, 1000, 100,000. Show how the mean-of-means, its standard deviation, and the confidence region depends on N.

```python
def normal_distr(R, N):
    mu = np.random.normal(size=(R,N)).mean(axis=1)
    print("Mean of means:", mu.mean())
    barx = mu.mean()
    std = np.sqrt(mu.var())
    print("Standard deviation:",std)
    x1 = np.percentile(mu, 97.5)
    x2 = np.percentile(mu, 2.5)

    outside = np.logical_or(mu < x2, mu > x1)
    print("confidence interval:", x2, x1)
    print("95% Confidence Region: ", outside.mean())
```

```python
normal_distr(R=1000,N=10)
```

```python
normal_distr(R=1000,N=1000)
```

```python
normal_distr(R=1000,N=100000)
```

Pick R and three N values.
Create R times a sample of N Pareto values, and take a mean of those.
Compute the mean-of-the-means, it's standard deviation, and 95% confidence region.
Repeat it with 3 different N-s.
Finally compare your results for normal and Pareto distributions.

```python
def pareto_distr(R, N):
    np.random.seed(123)
    mu = (np.random.pareto(a=0.5, size=(R, N))+1).mean(axis=1)
    print ("mean of means: ", mu.mean())
    stddev1 = mu.std()
    print ("standard dev: ", stddev1)
    x1 = np.percentile(mu, 97.5)
    x2 = np.percentile(mu, 2.5)

    outside = np.logical_or(mu < x2, mu > x1)
    print("confidence interval:", x2, x1)
    print("95% Confidence Region: ", outside.mean())
```

```python
pareto_distr(1000, N=1000)
```

```python
pareto_distr(1000, N=100)
```

```python
pareto_distr(1000, N=10)
```

#### Answer:
Compared to normal distributions, Pareto distributions have higher means and standard deviations. For k<=1, the denominator of expected value becomes very small and expected value shoots. The expected value for k<=1 doesn't have any significance as its value is very high and unstable unlike normal distribution as these distributions are highly skewed.
